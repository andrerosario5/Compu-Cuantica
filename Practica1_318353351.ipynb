## 1. Crea un circuito cuántico con un solo qubit y aplícale 5 compuertas diferentes:


### 1.1 Generando el circuito guardando el vector de estado

#### Importando Librerías de QisKit 

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex
from qiskit.visualization import plot_bloch_multivector
import numpy as np
```

#### Creando Circuito con un Qubit

```python
circ1 = QuantumCircuit(1)    # Creando circuito con un cubit 
psi_0 = Statevector(circ1)   # Estado despues de crear el circuito
```

#### Añadiendo la compuerta $X$

```python
circ1.x(0)                   # Apicar compuerta X al qubit 0
psi_1 = Statevector(circ1)   # Estado despues de aplicar la compuerta X -> |1>
display(circ1.draw('mpl'))   # Circuito con la primera compuerta
```

#### Añadiendo la compuerta $Z$

```python
circ1.z(0)                   # Apicar compuerta Z al qubit 
psi_2 = Statevector(circ1)   # Estado despues de aplicar la compuerta z -> -|1>
display(circ1.draw('mpl'))   # Circuito con la segunda compuerta
```

#### Añadiendo la compuerta $Y$

```python
circ1.y(0)                   # Apicar compuerta Y al qubit 
psi_3 = Statevector(circ1)   # Estado despues de aplicar la compuerta Y ->
display(circ1.draw('mpl'))   # Circuito con la tercera compuerta
```

#### Añadiendo la compuerta $H$

```python
circ1.h(0)                   # Apicar compuerta H al qubit 
psi_4 = Statevector(circ1)   # Estado despues de aplicar la compuerta H ->
display(circ1.draw('mpl'))   # Circuito con la cuarta compuerta
```

#### Añadiendo la compuerta $Rz()$

```python
circ1.rz(np.pi, 0)           # Apicar compuerta  Rz al qubit 
psi_5 = Statevector(circ1)   # Estado despues de aplicar la compuerta Rz ->
display(circ1.draw('mpl'))   # Circuito con la Quinta compuerta
```

#### Mostrando Estado Inicial (1)

```python
print("👉 Estado inicial:")
print("➤ en forma de vector:")
display(array_to_latex(psi_0))
print("➤ en forma de ket:")
display(psi_0.draw('latex'))
print()
```

#### Mostrando estado después de la compuerta $X$ (2)

```python
print("👉 Estado después de la compuerta X:")
print("➤ en forma de vector:")
display(array_to_latex(psi_1))
print("➤ en forma de ket:")
display(psi_1.draw('latex'))
print()
```

#### Mostrando estado después de la compuerta $Z$ (3)

```python
print("👉 Estado después de la compuerta Z:")
print("➤ en forma de vector:")
display(array_to_latex(psi_2))
print("➤ en forma de ket:")
display(psi_2.draw('latex'))
print()
```

#### Mostrando estado después de la compuerta $Y$ (4)

```python
print("👉 Estado después de la compuerta Y:")
print("➤ en forma de vector:")
display(array_to_latex(psi_3))
print("➤ en forma de ket:")
display(psi_3.draw('latex'))
print()
```

#### Mostrando estado después de la compuerta $H$ (5)

```python
print("👉 Estado después de la compuerta H:")
print("➤ en forma de vector:")
display(array_to_latex(psi_4))
print("➤ en forma de ket:")
display(psi_4.draw('latex'))
print()
```

#### Mostrando estado depues de la compuerta $Rz(\pi)$ (6) 

```python
print("👉 Estado después de la compuerta Rz:")
print("➤ en forma de vector:")
display(array_to_latex(psi_5))
print("➤ en forma de ket:")
display(psi_5.draw('latex'))
print()
```

#### Mostrando Resultado usando la Esfera de Bloch

```python
print()
print("🔵 Estado inicial:")
display(plot_bloch_multivector(psi_0))
# Primera compuerta
print("🔵 Estado después de aplicar la compuerta X:")
display(plot_bloch_multivector(psi_1))
#Segunda compuerta
print("🔵 Estado después de aplicar la compuerta Z:")
display(plot_bloch_multivector(psi_2))
#Tercera compuerta
print("🔵 Estado después de aplicar la compuerta Y:")
display(plot_bloch_multivector(psi_3))
#Cuarta compuerta
print("🔵 Estado después de aplicar la compuerta H:")
display(plot_bloch_multivector(psi_4))
#Quinta compuerta
print("🔵 Estado después de aplicar la compuerta de rotación a pi radianes Rz:")
display(plot_bloch_multivector(psi_5))
```

### 1.2 Explicación de cada compuerta

Este experimento muestra cómo cambia el estado de un qubit al aplicarle diferentes compuertas cuánticas. Inicialmente, el qubit está en el estado base, y conforme avanzamos, analizamos su transformación en la esfera de Bloch.

1. **Estado inicial \($\ket{0}$)**  
   - El qubit comienza en el estado base $\ket{0}$ = \begin{bmatrix}1 \\ 0\end{bmatrix} que se encuentra en el polo norte de la esfera de Bloch.

2. **Aplicación de la compuerta $X$**  
   - La compuerta $X$ es equivalente al $NOT$, intercambia $\ket{0}$y $\ket{1}$, por lo que ahora el estado del qubit es $\ket{1}$ = \begin{bmatrix}0 \\ 1\end{bmatrix}  moviéndose al polo sur de la esfera de Bloch.

3. **Aplicación de la compuerta $Z$**  
   - La compuerta Z introduce una fase de -1 en $\ket{1}$, convirtiéndolo en $-\ket{1}$. Aunque esto no cambia su posición en la esfera de Bloch, sí afecta su fase.

4. **Aplicación de la compuerta $Y$**  
   - La compuerta Y combina las compuertas $X$ y $Z$ y rota el estado en la esfera de Bloch. Transforma $\ket{1}$ en $i\ket{1}$, lo que equivale a una rotación de 180° alrededor del eje $Y$ posicionandolo nuevamente en polo norte de la esfera de Bloch.

5. **Aplicación de la compuerta Hadamard (H)**  
   - La compuerta H coloca el qubit en una superposición equilibrada:  
     $[
     H|0\rangle = \frac{\sqrt{2}i}{2}(\ket{0} + \ket{1})
     ]$  
     Esto sitúa el estado en el eje X de la esfera de Bloch, dividiendo la probabilidad de medir $\ket{0}$ o $\ket{1}$ en un 50%.

6. **Aplicación de la compuerta Rz(π)**  
   - La compuerta Rz(π) rota el estado alrededor del eje Z en $\pi$ radianes. Esto introduce una fase relativa sin afectar las probabilidades de medición.

Cada una de estas transformaciones se vieron en la esfera de Bloch, permitiéndome comprender cómo evoluciona el estado del qubit en el transcurso del circuito.

### 1.3 Expresión Matemática

#### Estado Inicial

$$\ket{\psi_3}=\ket{0}=\mymatrix{r}{ 1 \\ 0 }$$

#### Aplicación de la compuerta $X$

$$X\ket{0}=\mymatrix{rr}{ 0 & 1 \\ 1 & 0 }\mymatrix{r}{ 1 \\ 0 }=\mymatrix{r}{ 0 \\ 1 }=\ket{1}$$

#### Aplicación de la compuerta $Z$

$$Z\ket{1}=\mymatrix{rr}{ 1 & 0 \\ 0 & -1 }\mymatrix{r}{ 0 \\ 1 }=\mymatrix{r}{ 0 \\ -1 }=-\ket{1}$$

#### Aplicación de la compuerta $Y$

$$Y\ket{1} = \mymatrix{rr}{ 0 & -i \\ i & 0 }\mymatrix{r}{ 0 \\ 1 } = \mymatrix{r}{ -i \\ 0 } = -i\ket{0}$$

Entonces si es $-\ket{1}$:
$$\ket{\psi_3}=-(-i\ket{0})=i\ket{0}$$

#### Aplicación de la compuerta $H$

$$H\ket{0} = \frac{1}{\sqrt{2}} \mymatrix{rr}{ 1 & 1 \\ 1 & -1 }\mymatrix{r}{ 1 \\ 0 } = \frac{1}{\sqrt{2}} \mymatrix{r}{ 1 \\ 1 } = \frac{1}{\sqrt{2}} (\ket{0} + \ket{1})$$

Entonces si es $i\ket{0}$ :
$$\ket{\psi_4}=i*\frac{1}{\sqrt{2}} (\ket{0} + \ket{1})=\frac{i}{\sqrt{2}} (\ket{0} + \ket{1})$$

o bien puede representarse tambien:
$$\ket{\psi_4}=\frac{\sqrt{2}i}{2}(\ket{0} + \ket{1})=\frac{\sqrt{2}i}{2}\ket{0}+\frac{\sqrt{2}i}{2}\ket{1}$$

#### Aplicación de la compuerta $R_z(\phi)$

$$Rz(\pi)\ket{\psi_4} = \mymatrix{cc}{ e^{-i\frac{\pi}{2}} & 0 \\ 0 & e^{i\frac{\pi}{2}} }\mymatrix{r}{ \frac{i}{\sqrt{2}} \\ \frac{i}{\sqrt{2}} } = \mymatrix{rr}{ -i & 0 \\ 0 & i }\mymatrix{r}{ \frac{i}{\sqrt{2}} \\ \frac{i}{\sqrt{2}} } = \frac{1}{\sqrt{2}} \mymatrix{r}{ 1 \\ -1 } = \frac{1}{\sqrt{2}} (\ket{0} - \ket{1})$$

o bien puede representarse tambien:
$$\ket{\psi_4}=\frac{\sqrt{2}}{2}(\ket{0} - \ket{1})=\frac{\sqrt{2}}{2}\ket{0}-\frac{\sqrt{2}}{2}\ket{1}$$
