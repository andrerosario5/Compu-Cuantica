#### 1. Implementa el siguiente circuito cuántico con Qiskit y simúlalo. Recuerda que, por defecto, el estado inicial es $\ket{000}$, ¿cuál es el estado final? (mide los 3 qubits). El objetivo es averiguar qué hace el circuito cuántico.

<center>

##### Importando Librerías de Qiskit

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import array_to_latex
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Operator
from qiskit.providers.basic_provider import BasicSimulator
from qiskit.visualization import plot_histogram
from qiskit.primitives import StatevectorSampler
import numpy as np
```

##### Creando Circuito con el Estado Base $\ket{000}$

```python
circ = QuantumCircuit(3, 3)    # circuito con 3 qubits y 3 bits clásicos

circ.h(2)                   # aplicar compuerta H al qubit 2
circ.cx(1,2)                # aplicar compuerta CNOT
circ.tdg(2)                 # aplicar compuerta T Dagger al qbit 2
circ.cx(0,2)                # aplicar compuerta CNOT
circ.t(2)                   # aplicar compuerta T al qbit 2
circ.cx(1,2)                # aplicar compuerta CNOT
circ.tdg(2)                 # aplicar compuerta T Dagger al qbit 2
circ.cx(0,2)                # aplicar compuerta CNOT
circ.barrier()              #Barrera para mejorar la visualización del circuito
circ.t(1)                   # aplicar compuerta T al qbit 1
circ.t(2)                   # aplicar compuerta T al qbit 2
circ.cx(0,1)                # aplicar compuerta CNOT
circ.h(2)                   # aplicar compuerta H al qubit 2
circ.t(0)                   # aplicar compuerta T Dagger al qbit 0
circ.tdg(1)                 # aplicar compuerta T Dagger al qbit 1
circ.cx(0,1)                # aplicar compuerta CNOT
circ.barrier()              #Barrera para mejorar la visualización del circuito

circ.measure([0,1,2], [0,1,2])  # medir los 3 qubits en los 3 bits clásicos

display(circ.draw('mpl'))   # mostrar el circuito
```

Simulando los resultados

```python
simulator_basic = BasicSimulator()
job = simulator_basic.run(circ, shots=1024) ##Colapsa a un estado
result = job.result()
result

counts = result.get_counts()
counts

plot_histogram(counts, legend=["Estado 0"])
```
